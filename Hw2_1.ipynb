```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
URL = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
birth_rate = pd.read_html(URL, thousands='.', decimal=",", match='Коефіцієнт народжуваності в регіонах України')[0]
birth_rate
```

Вивести перші рядки таблиці за допомогою методу head

```python
birth_rate.head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
birth_rate.shape
```

Замініть у таблиці значення "-" на значення NaN

```python
birth_rate.replace(to_replace="—", value=np.nan, inplace=True)
birth_rate
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
birth_rate.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "-"

```python
birth_rate[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']] = birth_rate[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']].apply(pd.to_numeric)
birth_rate.dtypes
```

Порахуйте, яка частка пропущених даних міститься в кожній колонці (використовуйте методи isnull та sum)

```python
birth_rate.isnull().sum()/28*100
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
br = birth_rate.drop([birth_rate.shape[0]-1])
br
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
birth_rate.fillna(birth_rate.mean(numeric_only=True)).round(1)
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
birth_rate[ birth_rate['2019'] > birth_rate['2019'].mean()][['Регіон', '2019'] ]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
birth_rate.iloc[birth_rate['2014'].idxmax()]['Регіон']
```

Побудуйте стовпцеву діаграму народжуваності по регіонах у 2019 році

```python
plt.bar(birth_rate['Регіон'], birth_rate['2019'])
plt.title('Діаграма народжуваності по регіонах в 2019 році')
plt.tick_params(labelrotation = 90)
plt.show()
```
